```python
import time
import json
import pandas as pd
from kafka import KafkaConsumer
from pandas import DataFrame
from datetime import datetime
from jsonpath_ng import jsonpath, parse
from datetime import datetime
```

```python
brokers='localhost:9092'
topic='twitter'
sleep_time=5
offset='latest'
```

```python
consumer = KafkaConsumer(bootstrap_servers=brokers, auto_offset_reset=offset,consumer_timeout_ms=1000)
consumer.subscribe([topic])
```

```python
list_header = ['created_at','location','user','text']
mergedata = pd.DataFrame(data = [], columns=[list_header])
mergedata.to_csv('twitter.csv', mode='a', header=True)
```

```python
jsonpath_expression1 = parse('$.created_at')
jsonpath_expression2 = parse('$.user.location')
jsonpath_expression3 = parse('$.user.screen_name')
jsonpath_expression4 = parse('$.text')

master = ''
while(True):
    for message in consumer:
        json_data = json.loads(message.value)
#        print(json_data)
#        df1 = pd.DataFrame(json_data)
#        print(df1)
        for match in jsonpath_expression1.find(json_data):
#           print(f'{match.value}')
            created_at = match.value
            data1 = pd.DataFrame(data=[{created_at}],columns=['created_at'])
        for match2 in jsonpath_expression2.find(json_data):
#           print(f'{match2.value}')
            location = match2.value
            data2 = pd.DataFrame(data=[{location}],columns=['location'])
        for match3 in jsonpath_expression3.find(json_data):
#           print(f'{match3.value}')
            screen_name = match3.value
            data3 = pd.DataFrame(data=[{screen_name}],columns=['screen_name'])
        for match4 in jsonpath_expression4.find(json_data):
#           print(f'{match5.value}'
            text = match4.value
            data4 = pd.DataFrame(data=[{text}],columns=['text'])
        
        mergedata1 = data1.join(data2)

        mergedata2 = data3.join(data4)

        mergedata = mergedata1.join(mergedata2)
        print(mergedata)
        
        mergedata.to_csv('data'.csv',mode='a', header=False)
```
